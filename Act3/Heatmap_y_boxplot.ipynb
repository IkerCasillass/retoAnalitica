```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("clientes-centro-comercial.csv")
print(df.head(10))
```

```python
df = df.drop('IDcliente',axis=1)
df.describe()
```

```python
plt.clf()
fig, axs = plt.subplots(1,3, figsize = (12,4))


sns.boxplot(data=df, y = 'Edad', x = 'Genero', ax = axs[0])
sns.boxplot(data=df, y = 'Puntos en compras (1-100)',x = 'Genero', ax = axs[1])
sns.boxplot(data=df, y = 'Ingresos Anuales (k$)', x = 'Genero', ax = axs[2])
```

```python
fig = plt.figure()
sns.histplot(data=df, x = 'Edad', hue = 'Genero', bins=10, kde=True)

plt.xlabel("Edades")
plt.ylabel("Frecuencias")
plt.title("Histograma de Edad y Genero")
```

```python
corr = df.corr()
sns.heatmap(data=corr,cmap='gnuplot', annot = True)
```

```python
sns.pairplot(data=df, hue="Genero")
```
